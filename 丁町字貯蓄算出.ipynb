```python
import pandas as pd
import numpy as np
import geopandas as gpd
import os
```

```python
print('開始する都道府県コードを入力してください')
start_pref = input()
print(f'開始する都道府県コード : {start_pref}')

print('終了する都道府県コードを入力してください')
finish_pref = input()
print(f'終了する都道府県コード : {finish_pref}')

#丁町字ごとの貯蓄
def saving(gdf):
    columns_list = gdf.columns.to_list()
    columns_list = columns_list[12:31]
    saving_list = [1000000,1500000,2500000,3500000,4500000,5500000,6500000,7500000,8500000,9500000,11000000,13000000,15000000,17000000,19000000,22500000,27500000,35000000,40000000]
    
    #各階級ごとの金額合計のカラムを追加
    for i, j in zip(columns_list, saving_list):
        gdf[f'{i}_money'] = gdf[f'{i}'].astype(float)*j
        
    #金額合計カラムの総和をとり総戸建て数で割りsavingカラムに追加する
    gdf['total_saving'] = gdf.apply(lambda row: row.iloc[32:51].sum(), axis=1)
    gdf['saving'] = gdf['total_saving'] / gdf['（総数）世帯数'].astype(float)

%cd "C:/Users/imai_14/Documents/秋山研/卒論/負担率/貯蓄高階級別世帯数推計データ"
n = int(start_pref)
while n <= int(finish_pref):
    #ファイルを読み込み
    gdf = gpd.read_file(f'./{n}_saving.csv')
    
    #関数を用いてsavingカラムに貯蓄を追加
    #saving(gdf)
    saving(gdf)
    
    gdf.to_csv(f'./output/{n}_saving_output.csv', index=False)
    n += 1
```
