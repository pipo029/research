```python
import pandas as pd
import numpy as np
import geopandas as gpd
import os
```

```python
print('開始する都道府県コードを入力してください')
start_pref = input()
print(f'開始する都道府県コード : {start_pref}')

print('終了する都道府県コードを入力してください')
finish_pref = input()
print(f'終了する都道府県コード : {finish_pref}')

#丁町字ごとの貯蓄
def saving(df):
    columns_list = df.columns.to_list()
    columns_list = columns_list[12:31]
    saving_list = [1000000,1500000,2500000,3500000,4500000,5500000,6500000,7500000,8500000,9500000,11000000,13000000,15000000,17000000,19000000,22500000,27500000,35000000,40000000]
    
    #各階級ごとの金額合計のカラムを追加
    for i, j in zip(columns_list, saving_list):
        df[f'{i}_money'] = df[f'{i}'].astype(float)*j
        
    #金額合計カラムの総和をとり総戸建て数で割りsavingカラムに追加する
    df['total_saving'] = df.apply(lambda row: row.iloc[32:51].sum(), axis=1)
    df['saving'] = df['total_saving'] / df['（総数）世帯数'].astype(float)

%cd "G:/マイドライブ/akiyamalab/卒論/負担率総集編/"
n = int(start_pref)
while n <= int(finish_pref):
    #ファイルを読み込み
    saving_path = f'./data/貯蓄高階級別世帯数推計data/{n}_saving.csv'
    df = pd.read_csv(saving_path, encoding='shift-jis')
    
    #関数を用いてsavingカラムに貯蓄を追加
    #saving(df)
    saving(df)
    
    saving_output_path = f'./dev/貯蓄/{n}_saving_output.csv'
    df.to_csv(saving_output_path, encoding='shift-jis', index=False)
    n += 1
```
